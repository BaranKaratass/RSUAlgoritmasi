```python
import matplotlib.pyplot as plt
import time

class ChaosRNG:
    def __init__(self, seed=None):
        if seed is None:
            # Sistem zamanndan daha yksek hassasiyetli bir seed alyoruz
            self.x = (time.time_ns() % 1000000) / 1000000.0
        else:
            self.x = seed
        self.r = 3.999995  # Kaos katsays

    def generate(self):
        # Lojistik Harita Denklemi
        self.x = self.r * self.x * (1 - self.x)
        # 0-255 aras tam say retimi
        return int((self.x * 10**10) % 256)

# 1. Veri retimi (10,000 adet say)
rng = ChaosRNG()
data = [rng.generate() for _ in range(10000)]

# 2. Grselletirme
plt.figure(figsize=(12, 5))

# Sol Grafik: Histogram (Saylarn Dalm)
plt.subplot(1, 2, 1)
plt.hist(data, bins=64, color='skyblue', edgecolor='black')
plt.title("Saylarn Dalm (Histogram)")
plt.xlabel("retilen Say")
plt.ylabel("Frekans")

# Sa Grafik: Zaman Serisi (lk 100 Say arasndaki iliki)
plt.subplot(1, 2, 2)
plt.plot(data[:100], marker='o', linestyle='-', markersize=4, color='salmon')
plt.title("lk 100 Say Arasndaki liki (Kaos Testi)")
plt.xlabel("Sra")
plt.ylabel("Deer")

plt.tight_layout()
plt.show()
```

```python
import time

class ChaosRNG:
    def __init__(self, seed=None):
        # Yksek znrlkl zaman damgas ile gerek dnya entropisi
        self.x = (time.time_ns() % 1000000) / 1000000.0
        self.r = 3.999995

    def generate(self):
        self.x = self.r * self.x * (1 - self.x)
        return int((self.x * 10**10) % 256)

# Veri toplama
rng = ChaosRNG()
sayilar = [rng.generate() for _ in range(50)] # rnekleme iin 50 say

# Rapor Hazrlama
rapor_basligi = "KAOTK RNG GVENLK ANALZ RAPORU"
zaman_damgasi = time.strftime("%Y-%m-%d %H:%M:%S")

print(f"{'='*40}")
print(f"{rapor_basligi}")
print(f"Tarih: {zaman_damgasi}")
print(f"{'='*40}\n")

print(f"{'Sra':<10} | {'retilen Say (Byte)':<20} | {'Bit Deeri (Binary)':<15}")
print("-" * 55)

with open("analiz_raporu.txt", "w", encoding="utf-8") as f:
    f.write(f"{rapor_basligi}\n{zaman_damgasi}\n{'-'*40}\n")

    for i, sayi in enumerate(sayilar, 1):
        binary_hali = bin(sayi)[2:].zfill(8)
        satir = f"{i:<10} | {sayi:<20} | {binary_hali:<15}"
        print(satir)
        f.write(satir + "\n")

# Basit Bir Gvenlik statistii Ekleyelim
farklar = [abs(sayilar[i] - sayilar[i-1]) for i in range(1, len(sayilar))]
ortalama_fark = sum(farklar) / len(farklar)

analiz_notu = f"\nAnaliz Notu: Saylar aras ortalama deiim iddeti: {ortalama_fark:.2f}"
print(analiz_notu)
print(f"\n[!] Tm veriler 'analiz_raporu.txt' dosyasna kaydedildi.")
```
